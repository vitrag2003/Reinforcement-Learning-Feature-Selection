```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import *
# from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn import metrics
import seaborn as sns
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
df=pd.read_csv('main_dataset.csv')
```

```python
df
```

```python
# Separate features and target
X = df.drop(columns=["class"])  # Assuming "class" is the target variable
y = df["class"]
```

```python
# Select 75% of the most important features using RFE
num_features = int(0.75 * X.shape[1])
rfe = RFE(LogisticRegression(max_iter=1000), n_features_to_select=num_features)
rfe.fit(X, y)
```

```python
# Reduce dataset based on selected features
X_selected = X.loc[:, rfe.support_]
```

```python
# Save the reduced dataset
X_selected.to_csv("reduced_dataset.csv", index=False)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)
```

```python
# Train the model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')
```
